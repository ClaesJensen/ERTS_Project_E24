```python
import CustomEq as Eq
import math
import matplotlib.pyplot as plt
import numpy as np


fs = 48000
collection = Eq.FilterCollection([
    #Eq.PeakingEQ(Eq.FilterParameters.Q(-10, 1000, fs, 10)),
    Eq.Notch(Eq.FilterParameters.Q(0, 1000, fs, 10))
], fs)

f = 1000
frac = 22
time = [i/fs for i in range(fs)]
signal = np.array([math.sin(2*math.pi*f*t) * 2**frac for t in time])

fsignal = collection.processFp(signal, frac)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(signal[0:1000])
ax.plot(fsignal[0:1000])

collection.plotFreqResponse(True, True)
print(collection.coeffs(True, frac))
```
